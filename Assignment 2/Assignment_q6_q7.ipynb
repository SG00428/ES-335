Using Linear Regression for classification is not justified because Linear Regression is primarily designed for regression tasks, where the goal is to predict a continuous outcome variable based on one or more predictor variables.

In classification tasks, the goal is to predict the categorical class labels of instances based on their features. Linear Regression assumes that the relationship between the independent variables and the dependent variable is linear, and it tries to fit a linear function to the data. However, in classification problems, the relationship between the features and the class labels may not be linear, and the output is discrete rather than continuous.

Furthermore, Linear Regression can produce predictions outside the range of valid class labels, leading to inaccurate results.
